```python
import pandas as pd
import numpy as np
```

```python
url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

```python
chipo=pd.read_csv(url,sep='\t')
```

```python
c=chipo.groupby('item_name')                      #most ordered item in the dataset and number of items ordered.
c=c.sum()
c=c.sort_values(['quantity'],ascending=False)
c.head(1)
```

```python
total_items_orders=chipo.quantity.sum()
total_items_orders                                          #total ordered items
```

```python
chipo.item_name.value_counts().count()                  #number of different items sold
```

```python
c = chipo.groupby('choice_description').sum()
c = c.sort_values(['quantity'], ascending=False)            #most ordered item in choice description column.
c.head(1)
```

```python
revenue = (chipo['quantity']* chipo['item_price']).sum()
revenue                                                          #revenue
```

```python
orders = chipo.order_id.value_counts().count()          #orders made in period
orders
```
